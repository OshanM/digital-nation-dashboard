```python
import pandas as pd
import numpy as np
```

```python
#Read Sources cause all the data we need is there
sources = pd.read_csv('../../dashboard/sources.csv')
```

```python
#pivot per pillar and subpillar so we find out how many sources each country has for each pillar and subpillar
pillarpivot = pd.pivot_table(sources,values="Available",index=["Country Name"],columns=["Pillar"],aggfunc=np.sum)
subpillarpivot = pd.pivot_table(sources,values="Available",index=["Country Name"],columns=["Sub-Pillar"],aggfunc=np.sum)
pivot = pillarpivot.merge(subpillarpivot, on="Country Name")
```

```python
#Counting how many sources we have for pillar and subpillar
pillarcount = sources.groupby("Pillar")["Indicator"].nunique()
subpillarcount = sources.groupby("Sub-Pillar")["Indicator"].nunique()
#And joining those in series
counts = pillarcount.append(subpillarcount)
```

```python
#Pandas has this sweet option to get everything divided
pivot = pivot.div(counts,axis='columns')
#Uncomment this for % values
pivot = pivot.mul(100)
#uncomment to round
pivot = pivot.round(decimals=2)
pivot
```

```python
pivot.to_csv('../../dashboard/confidences.csv')
```
