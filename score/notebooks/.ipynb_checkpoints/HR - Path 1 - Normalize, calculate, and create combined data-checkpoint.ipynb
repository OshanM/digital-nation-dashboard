```python
### Average Scores by Pillars and Subpillars
```

```python
df = pd.read_csv('../../score/HR_country_scores/All.csv')
df = df[['Country Name','HRAssociation','agg_score','count_source']]
df
```

```python
# When normalizing country names, some entries in subpillar will be duplicated, to resolve this, the method is to multiple
# the agg_score with count_source and then divided them by the total number of source to get the average
df['total_agg'] = df['agg_score']*df['count_source']
df
```

```python
# Eliminate all duplicate Subpillar entries by aveerage them all
score_source = df.groupby(['Country Name','HRAssociation']).agg({'total_agg':'sum','count_source':'sum'}).reset_index()
score_source['final_score'] = score_source['total_agg']/score_source['count_source']
score_source.rename(columns = {'final_score': 'agg_score'}, inplace = True)
df = score_source
```

```python
#calculate the total average of sub-pillar score for each country
total_score =  df.groupby(['Country Name']).agg({'agg_score':'mean'}).reset_index()
total_score['HRAssociation'] = np.nan 
total_score = total_score[['Country Name','HRAssociation','agg_score']]
total_score.head(15)
```

```python
product = pd.concat([total_score, df], axis=0)
product.rename(columns={'agg_score':'Score'}, inplace=True)
product = product[['Country Name','HRAssociation','Score']]
```

```python
product.to_csv('../../score/HR_country_scores/Total_Scores.csv',index = False)
```
