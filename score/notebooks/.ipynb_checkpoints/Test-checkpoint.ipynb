```python
import pandas as pd
import numpy as np
import os 
```

```python
### Score Aggregating by Subpillars
```

```python
# get list of files in scores folder
scores = os.listdir('../subpillar_score/')
scores

df = pd.concat([pd.read_csv('../subpillar_score/{}'.format(s)) for s in scores])  
```

```python
df['Country Name'] = df['Country Name'].str.strip()
df['Country Name'] = df['Country Name'].str.strip('**')
df['Country Name'] = df['Country Name'].str.strip('*')
df['Country Name'] = df['Country Name'].str.strip('*')

# Replace as much code with country names as much as possible
df['Country Name'] = df['Country Name'].replace('UZ','Uzbekistan')
df['Country Name'] = df['Country Name'].replace('Vietnam','Viet Nam')
df['Country Name'] = df['Country Name'].replace('United States','United States of America')
df['Country Name'] = df['Country Name'].replace('UK','United Kingdom')
df['Country Name'] = df['Country Name'].replace('-',np.nan)
df['Country Name'] = df['Country Name'].replace('0.03',np.nan)
df = df.dropna()
```

```python
### Score Aggregating by Individual Indicators
```

```python
# get list of files in scores folder
scores = os.listdir('../indicator_scores/')
scores
df = pd.concat([pd.read_csv('../indicator_scores/{}'.format(s)) for s in scores])    
df = df[['Sub-Pillar','Indicator','Country Name','Year','data_col','new_rank_score','higher_is_better']]
df
```

```python
df['Country Name'] = df['Country Name'].str.strip()
df['Country Name'] = df['Country Name'].str.strip('**')
df['Country Name'] = df['Country Name'].str.strip('*')
df['Country Name'] = df['Country Name'].str.strip('*')

# Replace as much code with country names as much as possible
df['Country Name'] = df['Country Name'].replace('UZ','Uzbekistan')
df['Country Name'] = df['Country Name'].replace('Vietnam','Viet Nam')
df['Country Name'] = df['Country Name'].replace('United States','United States of America')
df['Country Name'] = df['Country Name'].replace('UK','United Kingdom')
df = df.dropna()
```

```python
df.insert(0,'Pillar','')
df.loc[(df['Sub-Pillar'] == 'Connectivity Technology') | (df['Sub-Pillar'] == 'Innovation Ecosystem'), 'Pillar'] = 'Infrastructure'
df.loc[(df['Sub-Pillar'] == 'Financing Incentives') | (df['Sub-Pillar'] == 'Technology Adoption') | (df['Sub-Pillar'] == 'Startup Environment'), 'Pillar'] = 'Business'
df.loc[(df['Sub-Pillar'] == 'Data Exchange') | (df['Sub-Pillar'] == 'Digital Legal Identity') | (df['Sub-Pillar'] == 'Digital Payments'), 'Pillar'] = 'Foundations'
df.loc[(df['Sub-Pillar'] == 'Capabilities') | (df['Sub-Pillar'] == 'Digital Public Services') | (df['Sub-Pillar'] == 'Funding and procurement') | (df['Sub-Pillar'] == 'Leadership and coordination'), 'Pillar'] = 'Government'
df.loc[(df['Sub-Pillar'] == 'Culture') | (df['Sub-Pillar'] == 'Digital Literacy Skills') | (df['Sub-Pillar'] == 'Digital Wellbeing')| (df['Sub-Pillar'] == 'Usage and ownership'), 'Pillar'] = 'People'
df.loc[(df['Sub-Pillar'] == 'Cybersecurity') | (df['Sub-Pillar'] == 'Ethical Standards'), 'Pillar'] = 'Regulation'
df.loc[(df['Sub-Pillar'] == 'Ambition'), 'Pillar'] = 'Strategy'

df
```

```python
# For all country
df = df[['Country Name','Year','Pillar','Sub-Pillar','Indicator','data_col','new_rank_score','higher_is_better']]
df.to_excel('../country_scores/All_Indicator.xlsx',index = False)
```
