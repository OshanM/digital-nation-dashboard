```python
sub_df = df.groupby(['Pillar','Sub-Pillar','Country Name']).agg({'new_rank_score':'mean','data_col':'count'}).reset_index()
sub_df.columns = ['Pillar','Sub-Pillar','Country Name','agg_score', 'count_source' ]
```

```python
max_number_sources = sub_df.describe()['count_source']['max']
sub_df['agg_score_wt'] = sub_df['agg_score']*(sub_df['count_source']/max_number_sources)
sub_df
sub_df
```

```python
#Calculate the average pillar score of each country
pillar_score = sub_df.groupby(['Country Name','Pillar']).agg({'agg_score':'mean'}).reset_index()
pillar_score.head(15)
```

```python
#calculate the total average of sub-pillar score for each country
total_score =  sub_df.groupby(['Country Name']).agg({'agg_score':'mean'}).reset_index()
total_score = total_score[['Country Name','agg_score']]
total_score.head(15)
```

```python
product = pd.merge(total_score, pillar_score, how="outer", on=["Country Name",'agg_score'])
product.insert(3,'Sub-Pillar','')
product
```

```python
product_total = pd.concat([product,sub_df], axis = 0).drop_duplicates().reset_index(drop=True)
product_total = product_total[['Country Name','agg_score','Pillar','Sub-Pillar']]
product_total.rename(columns = {'agg_score': 'Score'}, inplace = True)
product_total.rename(columns = {'Country Name': 'Country or Area'}, inplace = True)

product_total
```
