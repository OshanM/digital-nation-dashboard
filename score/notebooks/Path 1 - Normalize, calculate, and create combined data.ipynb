```python
### Average Scores by Pillars and Subpillars
```

```python
df = pd.read_csv('../country_scores/Combined.csv')
df
```

```python
# When normalizing country names, some sub-pillars will be duplicated, to resolve this, the method is to multiply
# the agg_score of the duplicate sub-pillars with count_source and then divided them by the total number of source to get the average
df['total_agg'] = df['agg_score']*df['count_source']
df
```

```python
# Eliminate all duplicate Subpillar entries by aveerage them all
df = df.groupby(['Country Name','Pillar','Sub-Pillar']).agg({'total_agg':'sum','count_source':'sum'}).reset_index()
df['final_score'] = df['total_agg']/df['count_source']
df.rename(columns = {'final_score': 'Score'}, inplace = True)
df
```

```python
#Calculate the average pillar score of each country
pillar_score = df.groupby(['Country Name','Pillar']).agg({'Score':'mean'}).reset_index()
pillar_score.head(15)
```

```python
#calculate the total average of sub-pillar score for each country
total_score =  df.groupby(['Country Name']).agg({'Score':'mean'}).reset_index()
total_score = total_score[['Country Name','Score']]
total_score.head(15)
```

```python
product = pd.merge(total_score, pillar_score, how="outer", on=["Country Name",'Score'])
product
```

```python
product_total = pd.merge(product, df, how="outer", on=["Country Name","Score","Pillar"])
product_total = product_total[['Country Name','Score','Pillar','Sub-Pillar']]
product_total = product_total[product_total['Country Name'] != 'Western Sahara']
product_total.sort_values(by=['Pillar'], inplace=True)

product_total
```

```python
product_total.to_csv('../country_scores/Total_Scores_V2.csv',index = False)
```
